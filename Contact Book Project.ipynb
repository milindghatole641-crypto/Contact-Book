```python
import os
import json
DATA_FILE="Contacts1.txt"

#Load Contacts From File.
def load_contacts():
    if not os.path.exists(DATA_FILE):
        return []
    else:
        with open(DATA_FILE,'r') as f:
            return json.load(f)

            

#Save Contact To File.
def save_contacts(contacts):
    with open(DATA_FILE,'w') as f:
        return json.dump(contacts,f,indent=4)

#Create New Contact.
def create_contacts():
    name = input("Enter Name:")
    phonenum = input("Enter Contact Number:")
    email = input("Enter E-Mail:")
    contact = {"name":name,"phone":phonenum,"email":email} 
    contacts = load_contacts()  #Load Existing Data.
    contacts.append(contact)    # Add New Contact.
    save_contacts(contacts)
    print("Contact Save Successfully!!")

#View Contacts Or Read Contacts.
def read_contacts():
    contacts = load_contacts()
    if not contacts:
        print("Contact Not Found")
    else:
        for idx,contact in enumerate(contacts,start=1):
             print(f"\nContact {idx} Details Are As Follows:")
             print("Name:",contact['name'])
             print("Contact No:",contact['phone'])
             print("E-Mail:",contact['email'])

#Update Contact.
def update_contacts():
    contacts = load_contacts()
    name = input("Enter The Name Of The Contact To Update")
    for contact in contacts:
        if contact['name'].lower() == name.lower():
            print("Contact Found.Leave Field Blank To Keep Current Value")
            new_phone = input(f"Enter New Phone({contact['phone']}):") or contact['phone']
            new_email = input(f"Enter New Email({contact['email']}):") or contact['email']
            contact['phone'] = new_phone
            contact['email'] = new_email
            save_contacts(contacts)
            print("Contact Updated Sucessfully...")
            return
            print("Contact Not Found...")

#Delete Contact.
def delete_contacts():
    contacts = load_contacts()
    name = input("Enter The Name Of The Contact To Delete")
    update_contacts = [c for c in contacts if c['name'].lower()!=name.lower()]
    if len(updated_contacts)==len(contacts):
        print("Contact Not Found.")
    else:
        save_contacts(updated_contacts)
        print("Contact Deleted Sucessfull...")

    
#Menu.
def menu():
    while True:
        print("---Contact Book---")
        print("1. Add Contacts")
        print("2. View Contacs")
        print("3. Update Contacts")
        print("4. Delete Contacts")
        print("5. Exit")
        ch=int(input("Enter The Choice:"))
        if ch==1:
            create_contacts()
        elif ch==2:
            read_contacts()
        elif ch==3:
            update_contacts()
        elif ch==4:
            delete_contacts()
        elif ch==5:
            print("Thanks!")
        break
    else:
        print("Invalid Choice . Please Try Again")

menu()
```
